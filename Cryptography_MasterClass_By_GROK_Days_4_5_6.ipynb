**Core Concept**  
A **hash** is a **fingerprint press**:  
- Same message ‚Üí **always same fingerprint**  
- Tiny change ‚Üí **completely different fingerprint** (avalanche)  
- Can‚Äôt go backward: no way to reconstruct message from fingerprint  
- Can‚Äôt find two messages with same fingerprint (collision-resistant)  

**SHA-256** works like a **meat grinder with 64 rounds**:  
1. Pad message to multiple of 512 bits  
2. Break into 512-bit blocks  
3. Start with 8 fixed "prime-root" constants (like salt from the universe)  
4. Each round: mix bits with rotations, XORs, additions modulo 2¬≥¬≤  
5. After 64 rounds: add back to constants ‚Üí final 256-bit fingerprint  

**Physical**: Imagine a **blender**.  
- Message = fruit chunks  
- Padding = fill to full jar  
- 64 rounds = blades spinning faster each time  
- Output = smoothie no one can un-blend

````python

def float_to_binary(num, precision=10):
    integer = int(num)
    fraction = num - integer

    # Convert integer part
    int_bin = bin(integer).replace("0b", "")

    # Convert fractional part
    frac_bin = ""
    for _ in range(precision):
        fraction *= 2
        bit = int(fraction)
        frac_bin += str(bit)
        fraction -= bit

    return f"{int_bin}.{frac_bin}"

print(float_to_binary(2**0.5, 10))
````

````python
# code to convery fractional part to hexadecimal
import math

def fractional_to_hex(num, bits=32):
    frac = num - int(num)
    value = int(frac * (1 << bits))  # multiply by 2^bits
    return hex(value)

print(fractional_to_hex(math.sqrt(19)))

# ________________________________________________________________________________________________________________________________________________________


def fractional_to_hex_general(num, bits=32):
    frac = num - int(num)
    scaled = int(frac * (1 << bits))
    return f"0x{scaled:0{bits//4}x}" # bits // 4 is used because 4 bits = 1 hex digit.

print(fractional_to_hex_general(2.0**0.5))  # sqrt(2)
print(fractional_to_hex_general(math.pi))   # œÄ fractional constant

````

````python
# Constants: First 32 bits of fractional parts of square roots of primes 2..311 - first 64 primes
K = [
    0x428a2f98, 0x71374491, 0xb5c0fbcf, 0xe9b5dba5, 0x3956c25b, 0x59f111f1, 0x923f82a4, 0xab1c5ed5,
    0xd807aa98, 0x12835b01, 0x243185be, 0x550c7dc3, 0x72be5d74, 0x80deb1fe, 0x9bdc06a7, 0xc19bf174,
    0xe49b69c1, 0xefbe4786, 0x0fc19dc6, 0x240ca1cc, 0x2de92c6f, 0x4a7484aa, 0x5cb0a9dc, 0x76f988da,
    0x983e5152, 0xa831c66d, 0xb00327c8, 0xbf597fc7, 0xc6e00bf3, 0xd5a79147, 0x06ca6351, 0x14292967,
    0x27b70a85, 0x2e1b2138, 0x4d2c6dfc, 0x53380d13, 0x650a7354, 0x766a0abb, 0x81c2c92e, 0x92722c85,
    0xa2bfe8a1, 0xa81a664b, 0xc24b8b70, 0xc76c51a3, 0xd192e819, 0xd6990624, 0xf40e3585, 0x106aa070,
    0x19a4c116, 0x1e376c08, 0x2748774c, 0x34b0bcb5, 0x391c0cb3, 0x4ed8aa4a, 0x5b9cca4f, 0x682e6ff3,
    0x748f82ee, 0x78a5636f, 0x84c87814, 0x8cc70208, 0x90befffa, 0xa4506ceb, 0xbef9a3f7, 0xc67178f2
]

# Initial hash values: Fractional parts of sqrt of primes 2..19
H = [
    0x6a09e667, 0xbb67ae85, 0x3c6ef372, 0xa54ff53a,
    0x510e527f, 0x9b05688c, 0x1f83d9ab, 0x5be0cd19
]

# _________________________________________________________________________________________________________________________________________________________

def rotr(x: int, y: int) -> int:
    """
    Rotate right (bits fall off right, land on left)
    Circular right shift. 
    Shifts x right by y bits and shifts x left by (32 - y) bits, then combines the two results using bitwise OR.
    """
    return ((x >> y) | (x << (32 - y))) & 0xFFFFFFFF

# here say we have 1001 & we want to shift by one bit.
# 1001 >> 1 = 0100
# 1001 << 3(4-1) = 1000
# now, 0100 | 1000 = 1100 which is the correct answer for rotating right by 1 bit.
# we do & 0xFFFFFFFF to ensure we only keep the lower 32 bits of the result, effectively simulating a 32-bit integer overflow.


# _________________________________________________________________________________________________________________________________________________________

def sha256(message: str) -> str:
    """
    Hand-rolled SHA-256. Input: string (UTF-8). Output: 64-char hex fingerprint.
    Matches: hashlib.sha256(message.encode('utf-8')).hexdigest()
    """
    # Step 1: Convert to UTF-8 bytes (reuse Day 3!)
    msg_bytes = str_to_utf8(message)
    
    # Step 2: Padding
    orig_len_bits = len(msg_bytes) * 8
    msg_bytes.append(0x80)  # append single '1' bit
    while (len(msg_bytes) % 64) != 56:
        msg_bytes.append(0x00)  # pad with zeros
    # Append original length as 64-bit big-endian
    for i in range(7, -1, -1):
        msg_bytes.append((orig_len_bits >> (i * 8)) & 0xFF)
    
    # Step 3: Process blocks
    h = H[:]  # copy initial state
    for block_start in range(0, len(msg_bytes), 64):
        block = msg_bytes[block_start:block_start+64]
        
        # Create message schedule W[0..63]
        w = [0] * 64
        for i in range(16):
            w[i] = bytes_to_int_tony(block[i*4:i*4+4])  # Tony's genius!
        for i in range(16, 64):
            s0 = rotr(w[i-15], 7) ^ rotr(w[i-15], 18) ^ (w[i-15] >> 3)
            s1 = rotr(w[i-2], 17) ^ rotr(w[i-2], 19) ^ (w[i-2] >> 10)
            w[i] = (w[i-16] + s0 + w[i-7] + s1) & 0xFFFFFFFF
        
        # Initialize working variables
        a, b, c, d, e, f, g, h_val = h
        
        # Main loop: 64 rounds
        for i in range(64):
            S1 = rotr(e, 6) ^ rotr(e, 11) ^ rotr(e, 25)
            ch = (e & f) ^ (~e & g)
            temp1 = (h_val + S1 + ch + K[i] + w[i]) & 0xFFFFFFFF
            S0 = rotr(a, 2) ^ rotr(a, 13) ^ rotr(a, 22)
            maj = (a & b) ^ (a & c) ^ (b & c)
            temp2 = (S0 + maj) & 0xFFFFFFFF
            
            h_val = g
            g = f
            f = e
            e = (d + temp1) & 0xFFFFFFFF
            d = c
            c = b
            b = a
            a = (temp1 + temp2) & 0xFFFFFFFF
        
        # Add to hash state
        h = [(x + y) & 0xFFFFFFFF for x, y in zip(h, [a, b, c, d, e, f, g, h_val])]
    
    # Final fingerprint
    return ''.join(f'{val:08x}' for val in h)
# _________________________________________________________________________________________________________________________________________________________
````

# *__Step-by-Step SHA-256 Hashing of "abc"__*

- SHA-256 is like a super secure blender for data‚Äîit takes any input (like the word "abc") and mixes it into a fixed-length "fingerprint" (a 256-bit number, shown as 64 hex characters). 
- The goal is to make this fingerprint unique: even a tiny change in the input flips about half the bits in the output (called the "avalanche effect," like a small snowball causing a big avalanche). 
- This makes it useful for passwords, file verification, or blockchain.

- We'll use the string "abc" as input. 
- I'll show binary (bits: 0s and 1s) for key parts, highlight changed or important bits with ` around them `(`e.g., 000`1`000 to spotlight the 1`), and include values in brackets like (hex: 0xabc, dec: 123) where helpful. 
- For complex operations like rotations or XOR, We'll break them down with examples from "abc" so you can visually see bits moving or flipping.

Terms to know (explained like you're 5):
- **`Bit`:**  A 0 or 1, the tiniest piece of data.
- **`Byte`:**  8 bits (e.g., 01100001 = 'a' in ASCII).
- **`Word`:**  Here, 32 bits (4 bytes).
- **`rotr(x, k)`:**  "Right rotate" ‚Äî imagine the 32 bits in a circle; shift them right by k positions, wrapping around. Why? To mix bits without losing any (like stirring soup).
- **`shr(x, k)`:**  Right shift ‚Äî push bits right by k, drop the ones that fall off, add 0s on left. Why? To divide by 2^k and focus on higher bits.
- **`^ (XOR)`:**  Bitwise "exclusive or" ‚Äî for each bit position, if the two inputs differ (one 0, one 1), output `1`; else 0. Why? Flips bits based on differences, creating unpredictability.
- **`& (AND), ~ (NOT)`:**  AND is 1 only if both 1; NOT flips all bits (0‚Üí1, 1‚Üí0).
- **`œÉ0 and œÉ1`:**  Special "scrambler" functions in the message schedule. They rotate/shift a word multiple ways, then XOR the results to diffuse bits (spread changes).
- **`S0 and S1`:**  Similar scramblers for the compression loop.
- **`Ch`:**  "Choose" ‚Äî picks bits from f or g based on e (like a switch).
- **`Maj`:**  "Majority" ‚Äî for each bit, if most of a/b/c are 1, output 1 (democratic vote).
- All additions are mod 2^32 (wrap around if over 4,294,967,295, causing "carries" that flip bits).

Since there are 64 rounds (lots of repetition), I'll detail the first few with bit highlights, then summarize the rest. Let's go!

### Step 2: Padding
Original length: 3 bytes = 24 bits (dec: 24).  
Why: 
- SHA-256 processes in 512-bit (64-byte) blocks.
- Padding makes the length a multiple of 512, marks the end of real data with a `1` bit, adds zeros for alignment, and appends the original bit length to prevent tricks like adding fake data (length-extension attacks). Leaves exactly 8 bytes for the length `(the last 64 bits)`.  
- Append 0x80 (binary: 10000000) ‚Äî the `1` says "data ends here," followed by 0s.  
  Original binary (first 24 bits): 011000010110001001100011  
  After append: ...01100011 `1`0000000 (highlighted the added `1` bit that changes an empty byte to 10000000).  
- Add zeros until length % 64 == 56 (here, 52 zero bytes, so total padded length 56 bytes before length).  
  Binary example (snippet): ...10000000 `00000000` x many (highlighted the zero padding as filler).  
- Append 64-bit length (big-endian: high bytes first): 00000000000000000000000000011000 (hex: 0x0000000000000018, with `11000` highlighted as the length bits).

Visual change: The `1` and length bits are new; zeros are neutral filler.

Why:

- Block 1: The full padded 64 bytes above.  
- Binary: Same as padded (no changes).  `(hex: 616263800000...00000018)  `

### Step 4: Message Schedule ‚Äî Expand Block to 64 Words
Why: 
- The 64-byte block is only 16 words (32-bit chunks). 
- We expand to 64 words for 64 mixing rounds. This diffuses bits: small input changes spread quickly via rotations/shifts/XOR/adds (avalanche).  
- First 16 words W`[`0-15`]`: Split block into 32-bit chunks (big-endian).  
- W`[`0`]`: 01100001011000100110001110000000 (hex: 0x61626380)  
- W`[`1-14`]`: 00000000000000000000000000000000 (hex: 0x0) ‚Äî all zeros from padding.  
- W`[`15`]`: 00000000000000000000000000011000 (hex: 0x18, with `11000` as length bits).  
- For W`[`16-63`]`: W`[`i`]` = W`[`i-16`]` + œÉ0(W`[`i-15`]`) + W`[`i-7`]` + œÉ1`(`W`[`i-2`])` mod 2^32.  
  **`œÉ0(x)`** = rotr`(`x,7`)` ^ rotr`(`x,18`)` ^ shr`(`x,3`)` ‚Äî scrambles by rotating/shifting then XORing.  
  **`œÉ1(x)`** = rotr`(`x,17`)` ^ rotr`(`x,19`)` ^ shr`(`x,10`)` ‚Äî similar, different amounts for variety.  

#### Example: W`[`17`]` (detailed bit changes).  
W`[`17`]` = W`[`1`]` (0) + œÉ0(W`[`2`]`=0) (0) + W`[`10`]` (0) + œÉ1(W`[`15`]`=0x18).  
Focus on œÉ1(W`[`15`]`):  
Original x = W`[`15`]`: 000000000000000000000000000`11`000 (highlighted the 1s at bits 3 and 4).  
- rotr17(x): Bits rotate right 17: the low 17 bits (`000000000000011000`) move to high. Result: 00000000000`11`000000000000000000 ( `11` highlighted as moved to bits 15-16). (hex: 0x30000)  
- rotr19(x): Rotate right 19: low 19 bits move to high. Result: 0000000000000`11`00000000000000000 ( `11` moved to bits 13-14). (hex: 0xc0000)  
- shr10(x): Shift right 10, drop low 10 bits (which had the 11000, so drops most 1s): 00000000000000000000000000000000 (all 0, highlighted no `1`s left).  
- XOR them: Bit by bit flip where odd number of 1s.  
  - At bit 16: rotr17 has 1, others 0 ‚Üí `1`  
  - Bit 15: rotr17 1, rotr19 0, shr 0 ‚Üí `1`

### 1) **What does ‚Äú448 mod 512‚Äù mean, and why 448?**

**Context:** SHA‚Äë256 processes the message in **512‚Äëbit blocks**. Before processing, it **pads** the message in a specific way:

1.  Append a single `1` bit.
2.  Append enough `0` bits so that the message length becomes **448 mod 512**.
3.  Append a **64‚Äëbit** number that encodes the original message length (before padding).

#### What is ‚Äú448 mod 512‚Äù?

*   ‚Äú`x ‚â° 448 (mod 512)`‚Äù means: when you divide `x` by 512, the remainder is 448.
*   In other words, make the message length land **448 bits** into a 512‚Äëbit block.

#### Why exactly **448**?

Because after step (2), we still need room to append the **64‚Äëbit length** in step (3).  
448 + 64 = **512**, so the block becomes perfectly full.

Think of a **512‚Äëbit ‚Äúbucket‚Äù**:

*   We fill it with the message, then put **a 1‚Äëbit marker** (‚Äúpadding starts here‚Äù), then zeros,
*   and we **reserve the last 64 bits** for the message length.  
    So we stop padding **at 448 bits**, leaving **exactly 64 bits** free at the end for the length.

**Why put the length at the end?**  
It‚Äôs part of the Merkle‚ÄìDamg√•rd construction. Encoding the original length at the end:

*   prevents certain structural ambiguities (e.g., two different messages with the same padded form),
*   and helps ensure security against some extension and collision tricks.

**What if the message is so long that adding `1` + zeros can‚Äôt land at 448?**  
Then we keep padding zeros so it spills into **another block**, and hit 448 in the **next** block, then append the 64‚Äëbit length there. Either way, the final padded message length is a **multiple of 512**.

***

### 2) **What are H0..H7? What do they store? Why primes? (Feynman style)**

#### The big picture:

SHA‚Äë256 keeps an **internal state**‚Äîthink of it like **8 small boxes** labeled:

    H0, H1, H2, H3, H4, H5, H6, H7

Each box holds a **32‚Äëbit number**.

*   At the **beginning**, these 8 numbers are set to special **initial constants** (IV = initial vector).
*   As each 512‚Äëbit block is processed, these numbers get **mixed** by bitwise operations (rotations, XORs, choices, majorities), absorbing the block‚Äôs data.
*   After all blocks, the final values in these 8 boxes are **concatenated** to form the **256‚Äëbit output**.

#### What do they store ‚Äúexactly‚Äù?

*   They store the **running digest state**‚Äîa **compressed summary** of everything processed so far.
*   Each round scrambles these values based on the current block‚Äôs words and constants so small changes in the message cause **large, unpredictable** changes in the state.

#### Why are **primes** involved?

Two places:

1.  **Initial H0..H7 constants**  
    They‚Äôre derived from the **fractional parts of the square roots of the first 8 primes** (2, 3, 5, 7, 11, 13, 17, 19).  
    Why?
    *   To provide **‚Äúnothing-up-my-sleeve‚Äù** constants‚Äînumbers that look random but are generated by a **simple, transparent rule** so no one can suspect hidden backdoors.
2.  **Round constants K\[0..63]**  
    These 64 constants come from the **fractional parts of the cube roots of the first 64 primes**.  
    Same reason: they appear random, are well-distributed, and are **publicly derivable**, making the design auditable.

**Feynman analogy:**  
Imagine you‚Äôre stirring a big pot of soup (the hash). The **H registers** are like the flavor levels (salty, sweet, sour‚Ä¶) at any moment. Each new ingredient (a 512‚Äëbit block) is added, and you stir using a precise stirring routine (the 64 rounds). The **prime-derived constants** are like your **standardized spice mix**‚Äîchosen in a way that **no chef** can secretly tweak it to hide a taste (backdoor). Everyone can see how the spice mix was derived, so it‚Äôs trusted.

***

### 3) **Why ‚Äú2^64 ‚àí 1‚Äù for input size? Why 2^64 at all?**

#### The ‚Äú64‚Äù comes from the **length field size**.

At the very end of padding, SHA‚Äë256 appends the original message length as a **64‚Äëbit unsigned integer** (big-endian).  
A **64‚Äëbit unsigned integer** can represent values from:

*   **0** up to **2^64 ‚àí 1**

So the **maximum number of bits** the algorithm can represent for the original length is **2^64 ‚àí 1**.  
That‚Äôs the **theoretical maximum input length in bits**.

#### Why not ‚Äú2^64‚Äù exactly?

Because if you count starting at 0, the **largest representable** number in 64 bits is **2^64 ‚àí 1**.  
Just like with 3 decimal digits, you can represent from 000 to 999 (which is 10^3 ‚àí 1).

#### What does that mean in size?

*   **2^64 ‚àí 1 bits ‚âà 1.84 √ó 10^19 bits**
*   That‚Äôs roughly **2.3 exabytes** of data (give or take conversion details).

**Practical note:**  
In real software, you can hash streams of arbitrary length because the data is processed block by block. You‚Äôre only limited by how the implementation tracks length and the time/storage to read the data. The **standard‚Äôs padding format** sets the theoretical bound via that **64‚Äëbit length field**.

***

### Quick recap (TL;DR)

*   **448 mod 512**: Stop padding at bit position 448 in the current 512‚Äëbit block so there‚Äôs room for the **64‚Äëbit length** at the end (448 + 64 = 512).
*   **H0..H7**: Eight 32‚Äëbit state words that hold the evolving hash state; initialized to constants derived from **primes** (sqrt/cuberoot fractional parts) to be transparent and ‚Äúnothing‚Äëup‚Äëmy‚Äësleeve‚Äù.
*   **2^64 ‚àí 1**: The length field is **64 bits**, so the maximum representable original message length is **2^64 ‚àí 1 bits**‚Äîthat‚Äôs why the input limit is stated that way.

## DAY 4.1 - "WHY DOES THE FINGERPRINT MACHINE WORK THIS WAY?"

**Goal**: Tony will understand **why** every single design choice in SHA-256 exists ‚Äî not just how, but the deep "why" behind primes, rotations, avalanche, and one-wayness.

---

**Core Concept (The Big "Why" of SHA-256)**

SHA-256 was designed in 2001 by the NSA (published as a public standard) to be a **cryptographic hash function** ‚Äî a fingerprint that is:

1. **Fast to compute** (easy to make smoothie)
2. **Impossible to reverse** (can't unblend fruit)
3. **Collision-resistant** (nearly impossible to find two messages with same fingerprint)
4. **Avalanche** (tiny change ‚Üí total chaos in output)

It belongs to the **Merkle-Damg√•rd** family, built on top of a **compression function** that mixes data in a way inspired by chaos theory and number theory.

Now, let‚Äôs go step by step ‚Äî **why each part exists**.

---

**The Deep "Why" Behind Each Step**

1. **Why use initial constants from square roots of primes?**  
   ‚Üí The 8 starting values (H0 to H7) are the **first 32 bits of the fractional parts of ‚àö2, ‚àö3, ‚àö5, ‚àö7, ‚àö11, ‚àö13, ‚àö17, ‚àö19**.  
   **Why?**  
   - Primes are "random-looking" in their distribution.  
   - Square roots of primes are **irrational** ‚Üí their fractional parts look like random noise.  
   - This breaks any hidden structure or symmetry that an attacker could exploit.  
   - If we used simple numbers like 0 or 1, someone could find patterns.  
   ‚Üí **Goal**: Start from a "nothing-up-my-sleeve" value ‚Äî numbers that no one can accuse the NSA of choosing secretly to hide a backdoor.

2. **Why the round constants K[0..63] from cube roots of primes?**  
   ‚Üí Same idea: first 32 bits of fractional parts of ‚àõ2, ‚àõ3, ..., ‚àõ311.  
   **Why per-round different constants?**  
   - Each round adds a unique "spice" so the mixing isn't regular or predictable.  
   - Prevents attacks where attacker cancels out rounds by symmetry.  
   ‚Üí **Goal**: Break periodicity. Make every round slightly different, like adding random seasoning.

3. **Why rotate bits (rotr)?**  
   ‚Üí Rotation mixes bits without losing information (unlike shifting, which throws bits away).  
   **Why different amounts (2,6,7,11,13,17,18,19,22,25)?**  
   - These numbers are **carefully chosen** so that after many rounds, every input bit affects every output bit ‚Äî **diffusion**.  
   - They are not multiples of each other ‚Üí no alignment ‚Üí maximum chaos.  
   ‚Üí **Physical**: Imagine stirring paint. Rotate by 10¬∞ each time ‚Üí eventually fully mixed.

4. **Why the functions Ch (choice) and Maj (majority)?**  
   - **Ch(e,f,g) = (e & f) ^ (~e & g)** ‚Üí "if e is 1, choose f, else choose g"  
     ‚Üí Acts like a **multiplexer** ‚Äî nonlinear switch.  
   - **Maj(a,b,c) = (a & b) ^ (a & c) ^ (b & c)** ‚Üí "vote": whichever value appears twice wins  
     ‚Üí Another nonlinear gate.  
   **Why nonlinear?**  
   - Linear functions (only XOR, add) are easy to reverse or solve.  
   - Nonlinear = attacker can't use algebra to undo steps.  
   ‚Üí **Goal**: Make reversal mathematically hard.

5. **Why œÉ0 and œÉ1 in message schedule?**  
   ‚Üí They expand 16 words ‚Üí 64 using rotations and XORs.  
   **Why?**  
   - Makes each output word depend on many input words ‚Üí **diffusion across message**.  
   - Even if message is short, it gets stretched and mixed heavily.

6. **Why add modulo 2¬≥¬≤ (32-bit wrap-around)?**  
   ‚Üí Prevents carrying over bits in a predictable way.  
   ‚Üí Creates **nonlinearity** through overflow.  
   ‚Üí Makes arithmetic attacks harder.

7. **Why 64 rounds?**  
   ‚Üí Enough for full diffusion: every bit affects every other bit many times over.  
   ‚Üí Fewer ‚Üí possible shortcut attacks. More ‚Üí slower, no extra security.

8. **Why padding with 0x80 + length?**  
   ‚Üí Ensures different length messages don‚Äôt collide after padding.  
   ‚Üí The "1" bit (0x80) marks end of real message ‚Üí prevents ambiguity.  
   ‚Üí Length at end ‚Üí even if you add zeros, different lengths ‚Üí different hashes (length extension attack defense in Merkle-Damg√•rd).

**Summary of All the "Whys"**:
| Part                  | Why It Exists                                      | Security Goal                 |
|-----------------------|----------------------------------------------------|-------------------------------|
| Prime square roots    | Nothing-up-my-sleeve numbers                       | No hidden patterns/backdoors  |
| Cube roots in K       | Break symmetry across rounds                       | Prevent round cancellation    |

## Modulo n addition


**Modulo n** means **the remainder after dividing by n**.

You can think of it as **counting in a circle of size n**:

* When you reach `n`, you wrap back to `0`.

Example:

* `7 mod 5 = 2`
* `3 + 4 = 7 ‚Üí 7 mod 5 = 2`

**In one line:**

> Modulo n keeps only what‚Äôs left after division and wraps numbers around at n.


### Simple number example

Let‚Äôs say **n = 5**:

| Number | Divide by 5 | Remainder (mod 5) |
| ------ | ----------- | ----------------- |
| 0      | 0           | 0                 |
| 3      | 0           | 3                 |
| 5      | 1           | 0                 |
| 7      | 1           | 2                 |
| 12     | 2           | 2                 |

So:

```
7 mod 5 = 2
```

---


````python
def mod_n_addition(*args, mod_n):
    """
    Perform modulo-n addition on a variable number of arguments.

    This function computes the sum of all positional arguments and returns
    the result modulo `mod_n`.

    Parameters
    ----------
    *args : int
        One or more integers to be added together.
    mod_n : int
        The modulus to apply. Must be a positive, non-zero integer.

    Returns
    -------
    int
        The result of (sum(args) % mod_n).

    Raises
    ------
    ValueError
        If no positional arguments are provided.
        If mod_n is less than or equal to zero.
    TypeError
        If any of the arguments or mod_n are not integers.

    Examples
    --------
    >>> mod_n_addition(3, 5, 7, mod_n=10)
    5

    >>> mod_n_addition(1, 2, 3, 4, mod_n=3)
    1
    """
    # Validate mod_n
    if not isinstance(mod_n, int):
        raise TypeError("mod_n must be an integer.")
    if mod_n <= 0:
        raise ValueError("mod_n must be a positive, non-zero integer.")

    # Validate args
    if not args:
        raise ValueError("At least one value must be provided in *args.")

    for i, value in enumerate(args):
        if not isinstance(value, int):
            raise TypeError(f"Argument at position {i} is not an integer: {value}")

    # Perform modulo-n addition
    total = sum(args)
    return total % mod_n
````
